## Binary Search Trees: Depth-First Traversal (Recursive)

```python
class BinarySearchTree:
    def __init__(self, in_list):
        self.binary_tree = []
        for x in in_list:
            self.add_node(x)

    def print_tree(self):
        print(self.binary_tree)
        return 0

    def add_node(self, value):
        i = 0
        node_items = {"data": value, "leftNode": None, "rightNode": None}
        node_position = len(self.binary_tree)
        while i < node_position:
            cur_value, left_node, right_node = self.binary_tree[i].values()
            if cur_value > value:
                if left_node is not None:
                    i = left_node
                else:
                    self.binary_tree[i]["leftNode"] = node_position
                    break
            else:
                if right_node is not None:
                    i = right_node
                else:
                    self.binary_tree[i]["rightNode"] = node_position
                    break
        self.binary_tree.append(node_items)
        return 0

    def add_nodes(self, values):
        for value in values:
            self.add_node(value)

    def traverse_DF_preorder_recursive(self):
        def preorder_node(node):
            data, left_node, right_node = self.binary_tree[node].values()
            print(data, end=", ")
            if left_node is not None:
                preorder_node(left_node)
            if right_node is not None:
                preorder_node(right_node)

        preorder_node(0)
        print()

    def traverse_DF_inorder_recursive(self):
        def inorder_node(node):
            data, left_node, right_node = self.binary_tree[node].values()
            if left_node is not None:
                inorder_node(left_node)
            print(data, end=", ")
            if right_node is not None:
                inorder_node(right_node)

        inorder_node(0)
        print()

    def traverse_DF_postorder_recursive(self):
        def postorder_node(node):
            data, left_node, right_node = self.binary_tree[node].values()
            if left_node is not None:
                postorder_node(left_node)
            if right_node is not None:
                postorder_node(right_node)
            print(data, end=", ")

        postorder_node(0)
        print()


class StackEmptyError(IndexError):
    pass


class StackList:
    def __init__(self):
        self.stack_list = []

    def push(self, value):
        self.stack_list.append(value)

    def pop(self):
        try:
            return self.stack_list.pop()
        except IndexError:
            raise StackEmptyError


if __name__ == "__main__":
    binary_tree1 = BinarySearchTree([12, 16, 2])
    binary_tree1.add_node(25)
    binary_tree1.add_node(3)
    binary_tree1.add_nodes([18, 6, 47])
    # print(binary_tree1.binary_tree)
    # binary_tree1.print_tree()
    print("Pre-order Depth-First Traversal - Recursive: ")
    binary_tree1.traverse_DF_preorder_recursive()
    print("In-order Depth-First Traversal - Recursive: ")
    binary_tree1.traverse_DF_inorder_recursive()
    print("Post-order Depth-First Traversal - Recursive: ")
    binary_tree1.traverse_DF_postorder_recursive()
```
